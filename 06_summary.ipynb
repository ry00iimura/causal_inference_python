# **本セミナーのまとめ**
- 統計モデリングとは何か？
  - データの生成過程に関して，確率的な仮説を置いて解析をする
  - データの理解を通して，予測や変化点検知などの多くの応用が生まれる
- ベイズ統計の基本
  - ベイズ統計は「モデル設計 x 推論計算」
  - モデル設計では，ガウス分布などの確率分布を組み合わせてアイデアを導入する
  - 推論計算では，PyMC3のMCMCなどのアルゴリズムを用いて近似的に結果を得る
